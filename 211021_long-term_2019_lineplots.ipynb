# Understanding the relationship(s) between water chemistry variables within and between flood events from the Burdekin River

**Methods:**

1. Load in required packages 
2. Load in master data file 
3. Long-term variable plots 
4. 2019 flood event plot

## Load in packages

```python
import pandas as pd

from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (16,8)
%matplotlib inline
```

## Load in master data file 

```python
# Load in df 
df = pd.read_csv(
        "211021_masterdf.csv",
        sep=r",",
        engine="c",
        header=0,
        na_filter=False,
        low_memory=False,
    )
# Adjust time and date column 
df['Aggregated Date/Time']= pd.to_datetime(df['Aggregated Date/Time'])
```

Check the dataframe

```python
df.head()
```

Remove spaces in names for easy manipulation

```python
df.columns = df.columns.str.replace(' ','_')
df.columns = df.columns.str.replace('/','_')
```

## Long-term variable plots

```python
# Assign each Site_Name to a new dataframe to plot: 
print(df['Site_Name'].unique())
rivermouth = df[(df.Site_Name == "Burdekin river rivermouth")]
offshore = df[(df.Site_Name == "Burdekin river offshore")]
tink = df[(df.Site_Name == "Tink shoal")]
wilson = df[(df.Site_Name == "Wilson Shoal")]
morinda = df[(df.Site_Name == "Morinda Shoal")]
old = df[(df.Site_Name == "Old Reef")]
stanley = df[(df.Site_Name == "Stanley Reef")]
bowden = df[(df.Site_Name == "Bowden Reef")]
darley = df[(df.Site_Name == "Darley Reef")]
broadhurst = df[(df.Site_Name == "(Big) Broadhurst Reef (No. 1)")]
charity = df[(df.Site_Name == "Charity reef")]
shell = df[(df.Site_Name == "Shell Reef")]
faith = df[(df.Site_Name == "Faith Reef")]
shrimp = df[(df.Site_Name == "Shrimp Reef")]
hope = df[(df.Site_Name == "Hope Reef")]
UN_19_280 = df[(df.Site_Name == "U/N Reef 19-280")]
viper = df[(df.Site_Name == "Viper Reef")]
UN_18_132 = df[(df.Site_Name == "U/N Reef 18-132")]
tiger = df[(df.Site_Name == "Tiger Reef")]
cup = df[(df.Site_Name == "Cup Reef (North)")]
leopard = df[(df.Site_Name == "Leopard reef")]
jacqueline = df[(df.Site_Name == "Jacqueline Reef")]

# make figure 
fig, ax = plt.subplots(figsize = (16, 8))

# Plot: 
plt.plot('Aggregated_Date_Time', 'omega', data = rivermouth, color = '#d00000', linewidth = 1, label = 'Rivermouth')
plt.plot('Aggregated_Date_Time', 'omega', data = offshore, color = '#FF0000', linewidth = 1, label = 'Offshore')
plt.plot('Aggregated_Date_Time', 'omega', data = tink, color = '#FFA200', linewidth = 1, label = 'Tink Shoal')
plt.plot('Aggregated_Date_Time', 'omega', data = wilson, color = '#FFAE00', linewidth = 1, label = 'Wilson Shoal')
plt.plot('Aggregated_Date_Time', 'omega', data = morinda, color = '#FFB900', linewidth = 1, label = 'Morinda Shoal')
plt.plot('Aggregated_Date_Time', 'omega', data = old, color = '#FFC500', linewidth = 1, label = 'Old Reef')
plt.plot('Aggregated_Date_Time', 'omega', data = stanley, color = '#FFD000', linewidth = 1, label = 'Stanley Reef')
plt.plot('Aggregated_Date_Time', 'omega', data = bowden, color = '#FFDC00', linewidth = 1, label = 'Bowden Reef')
plt.plot('Aggregated_Date_Time', 'omega', data = darley, color = '#FFE700', linewidth = 1, label = 'Darley Reef')
plt.plot('Aggregated_Date_Time', 'omega', data = broadhurst, color = '#FFF300', linewidth = 1, label = '(Big) Broadhurst Reef (No. 1)')
plt.plot('Aggregated_Date_Time', 'omega', data = charity, color = '#FFFF00', linewidth = 1, label = 'Charity Reef')
plt.plot('Aggregated_Date_Time', 'omega', data = shell, color = '#E7F317', linewidth = 1, label = 'Shell Reef')
plt.plot('Aggregated_Date_Time', 'omega', data = faith, color = '#D0E72E', linewidth = 1, label = 'Faith Reef')
plt.plot('Aggregated_Date_Time', 'omega', data = shrimp, color = '#B9DC45', linewidth = 1, label = 'Shrimp Reef')
plt.plot('Aggregated_Date_Time', 'omega', data = hope, color = '#A2D05C', linewidth = 1, label = 'Hope Reef')
plt.plot('Aggregated_Date_Time', 'omega', data = UN_19_280, color = '#8BC573', linewidth = 1, label = 'U/N 19-280')
plt.plot('Aggregated_Date_Time', 'omega', data = viper, color = '#73B98B', linewidth = 1, label = 'Viper Reef')
plt.plot('Aggregated_Date_Time', 'omega', data = UN_18_132, color = '#5CAEA2', linewidth = 1, label = 'U/N 18-132')
plt.plot('Aggregated_Date_Time', 'omega', data = tiger, color = '#45A2B9', linewidth = 1, label = 'Tiger Reef')
plt.plot('Aggregated_Date_Time', 'omega', data = cup, color = '#2E97D0', linewidth = 1, label = 'Cup Reef (north)')
plt.plot('Aggregated_Date_Time', 'omega', data = leopard, color = '#178BE7', linewidth = 1, label = 'Leopard Reef')
plt.plot('Aggregated_Date_Time', 'omega', data = jacqueline, color = '#0080FF', linewidth = 1, label = 'Jacqueline Reef')

plt.xlabel("Year", size = 15)
plt.ylabel("Aragonite Saturation State (Î©)", size= 15)
plt.title("Long-term Aragonite Saturation State (Î©)", size = 20)

# add horizontal lines for thresholds 
plt.axhline(y=2.82, color='k', linestyle='--', label = 'Î© Threshold: 2.82')


plt.legend(ncol = 3, loc = 'best', frameon = True, prop = {"size":11})
```
