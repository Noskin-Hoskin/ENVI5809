**Plot analysis:**

From these plots it is clear that all variables exhibit some degree of spatial and temporal varaibility. There is a clear seasonal fluctation in aragonite saturation state and pH, where values are higher in summer/autumn and lower in winter/spring. Both pH and aragnite saturation state follow a similar temporal trend. Salinity also exhibits some degree of temporal variability with clear troughs in late summer/autumn. While coral symbiont chlorophyll exhibits temporal variability, there is no clear pattern apparent. 

For all variables, there is a spatial relationship between HYDRO and BGC variables and their location relative to the rivermouth. Sites closer to the rivermouth (colours on the red-end of the spectrum) experience a greater deal of variability compared to sites further from the rivermouth (colours on the blue-end of the spectrum). This spatial trend is less apparent for coral symbiont chlorophyll compared to aragonite saturation state, pH, and salinity. 

From the plots of aragonite saturation state and pH, five potential flood events can be idenified. These occur in early 2011, early-mid 2012, mid 2017, early-mid 2018, and early 2019. There appears to be an inverse relationship between pH and aragonite saturation state both during the flood events and during regular seasonal variations. There is a direct relationship between changes in salinity and aragonite saturation state, and an inverse relationship between salinity and pH. The same five major flood events can be identified from the salinity data, alongside four additional smaller events that may be considered seasonal flucatuations as they occur annually during late-summer/early-autumn.

**Plot analysis:**

From these plots it is clear that all variables experience some variability during the 2019 flood event. It is clear that a flood discharge from the Burdekin River occurs around ~ 28th January and contributes to a rise in ocean pH and coral symbiont chlorophyll values and a fall in aragonite saturation state and salinity values. The peak of the flood disharge event appears to occur around ~5th February. Aragonite saturation state and pH exhibit the same inverse relationship, and both salinity and aragonite saturation state follow the same temporal pattern as expected. A clearer temporal trend for coral symbiont chlorophyll content can be seen as values rise around the flood discharge event. For araognitie saturation state, pH, and salinity, there is a lag between the flood discharge event at the rivermouth (colours on the red-end of the spectrum) and when it reaches sites towards the mid and outer-reef (sites on the green and blue-end of the spectrum). All HYDRO and BGC variables exhibit more variation closer to the rivermouth than sites further from the rivermouth.
